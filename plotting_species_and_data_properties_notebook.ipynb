```python
import scipy.stats as stats

fig = plt.figure(figsize=(10,10), tight_layout=True)
f = sns.jointplot(data=df_supermatrices,ylim=(0,1.1),x="No. of alignment sites",y="Prop. of rejected symmetry tests", kind="reg", color="purple")
r, p = stats.pearsonr(x=df_supermatrices["No. of alignment sites"], y=df_supermatrices["Prop. of rejected symmetry tests"])
phantom, = f.ax_joint.plot([], [], linestyle="", alpha=0)
ci =stats.pearsonr(x=df_supermatrices["No. of alignment sites"], y=df_supermatrices["Prop. of rejected symmetry tests"]).confidence_interval()
f.ax_joint.legend([phantom],['Pearson r: r={:5.5f}, p={:5.5f}'.format(r,p)])

plt.savefig("sites_vs_symmetry_prop.png")
plt.savefig("sites_vs_symmetry_prop.pdf")
plt.savefig("sites_vs_symmetry_prop.svg")
print(ci)
```
